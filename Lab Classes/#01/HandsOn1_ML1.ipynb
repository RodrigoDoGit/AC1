### Build and Evaluate a Machine Learning Model

Your task is to use the Hotel Reservations data set to build a decision tree to predict whether a reservation will be cancelled.

Load the modules sklearn and graphviz

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

Drop the column Booking_ID

```python
df.drop(columns='Booking_ID',inplace=True)
```

Map the values of booking_status canceled to 1 and not_canceled to 0

```python
df.booking_status= df['booking_status'].map({'Canceled': 1, 'Not_Canceled': 0})
```

Select the numeric columns

```python
df1=df.select_dtypes(include=np.number)
#df1.dtypes
```

Save in different variables the predictors and the target (booking_status)

```python
X = df1.drop(['booking_status'], axis = 1)
y = df1['booking_status']
#X.shape
#X.dtypes
```

Split the data in training and test with a split of 70%/30%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Setup the decision tree algorithm with maxdepth=3 and min_sample_split=5

```python
dt = DecisionTreeClassifier(max_depth=3, min_samples_split=5)
```

With the above setup, train a decision tree model

```python
dt.fit(X_train, y_train)
```

With the obtained model, make the predictions on the test set

```python
y_pred = dt.predict(X_test)
#print(classification_report(y_test, y_pred))
```

What is the accuracy of your model on the test set?

```python
accuracy=dt.score(X_test,y_test)
print("Accuracy = ",accuracy)
```

Plot the decision tree

```python
# Export decision tree to DOT format
dot_data = export_graphviz(dt, out_file=None, 
                           feature_names=X.columns,  
                           class_names=['not canceled','canceled'],
                           filled=True, rounded=True,  
                           special_characters=True)

# Plot decision tree
graph = graphviz.Source(dot_data)
graph.render("decision_tree.pdf")  # save to file
graph.view()  # open in viewer
```
